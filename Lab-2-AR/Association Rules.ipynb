```python
# Given dictionary of sequences and a threshold, eliminate all the items of the dictionary having support less than the threshold
# the support is calculated as frequency of item/numItems
# Step 0 - filtering single sequences

def filter_under_threshold(dict_occurrences, thresh = 2):
    to_rem_el = []
    for key in dict_occurrences.keys():
        if dict_occurrences[key]/numItems < thresh:
            to_rem_el.append(key)
            
    for el in to_rem_el:
        del dict_occurrences[el]
    
    return dict_occurrences
```

```python
#count number of occurences per itemSet 
#we create a dict with itemSet: occurences (Ck)

def get_ck_possible_itemset(dict_freq_items, receipts, n_plet):
    if n_plet < 1:
        print("Error: n_plet must be at least 1, retry")
        
    new_frequent_dict = {}
    
    if n_plet == 1:
        #skip the intersection
        for receipt in receipts:
            for el in receipt:
                if new_frequent_dict.get(el) is None:
                    new_frequent_dict[el] = 1
                else:
                    new_frequent_dict[el] += 1
                    
    # Case n_plet >= 2 - must compute subsequences
    else:
        # Get frequent itemset as list, from the dictionary
        
        if n_plet >=3:
            # Check if my dictionary has sequences of at list 2 lements. 
            # In this case, I need to return the single elements
            freq_items = set(itertools.chain(*dict_freq_items.keys()))
        else:
            freq_items = set(dict_freq_items.keys())

        # Start to compute the n_plets
        
        # 0 - initialize dictionary for frequent sublists at 0
        for receipt in receipts:
            # 1- get only the remained frequent items from the receipt I am iterating
            marginal_elements = freq_items.intersection(set(receipt))
            
            # 2- compute all possible subsequences of the marginal elements and save in new dict
            marginal_sequences = itertools.combinations(sorted(marginal_elements), n_plet) #TODO
            
            for seq in marginal_sequences:
                if new_frequent_dict.get(seq) is None:
                    new_frequent_dict[seq] = 1
                else:
                    new_frequent_dict[seq] += 1

    return new_frequent_dict
```
